1. Complete all exersices with Lynx-Rabbits dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'http://scipy-lectures.org/_downloads/populations.txt'
data = np.loadtxt(url, delimiter='\t', skiprows=1)
```

```python
data = data.astype(int)
data
```

```python
years = data[:,0]
hare = data[:,1]
lynx = data[:,2]
carrot = data[:,3]
```

Plot the populations for the years in the period. Add legend and axis labels to the plot.

```python
fig, ax = plt.subplots()

ax.plot(years, hare, '-o', label='Hare', c='lightgreen', linewidth=3)
ax.plot(years, lynx, '-o', label='Lynx', c='lightblue', linewidth=3)

fig.set_figwidth(18)    
fig.set_figheight(6)

plt.xticks(range(min(years), max(years) + 1))
plt.xlabel('Year')
plt.ylabel('Population')

plt.legend()
plt.show() 
```

2. Plot a graph of total explained variance vs number of principal components.

```python
from sklearn.datasets import fetch_openml
digits = fetch_openml('mnist_784')
images = digits.data.reshape(-1, 28, 28)
```

```python
i = 2209

plt.imshow(images[i], cmap='gray')
plt.title('This is {}'.format(digits.target[i]))
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=784)
pca.fit_transform(digits.data.astype(float))
var = []
num = 0

for v in pca.explained_variance_ratio_:
    var.append(num + v)
    num = var[-1]

var = np.asarray(var) 
```

```python
fig, ax = plt.subplots()

ax.plot(var*100, c = 'red', linewidth=3)

fig.set_figwidth(18)    
fig.set_figheight(6) 

plt.xlabel('The number of principal components')
plt.ylabel('Total explained variance')
plt.show()
```

3. Find a number of principal components for 99%, 95%, 90%, and 85% of explained variance.

```python
disp = np.zeros(5)  

disp[0] = np.argmin(np.abs(var * 100 - 85))
disp[1] = np.argmin(np.abs(var * 100 - 90))
disp[2] = np.argmin(np.abs(var * 100 - 95))
disp[3] = np.argmin(np.abs(var * 100 - 99))
disp[4] = np.argmin(np.abs(var * 100 - 100))

print('The number of principal components for 85% is:\t', int(disp[0]))
print('The number of principal components for 90% is:\t', int(disp[1]))
print('The number of principal components for 95% is:\t', int(disp[2]))
print('The number of principal components for 99% is:\t', int(disp[3]))
print('The number of principal components for original one is:\t', int(disp[4]))
```

4. Plot an original image and restored images from 99%, 95%, 90%, and 85% of explained variance.

```python
x = 2209

digits.data.astype(float)
images = digits.data.reshape(-1, 28, 28)

fig, ax = plt.subplots(1, 5)

for i in range(4):
    pca = PCA(n_components=int(disp[i]))
    X_transformed = pca.fit_transform(digits.data.astype(float))
    X_restored = pca.inverse_transform(X_transformed)
    ax[i].imshow(X_restored[x].reshape(28, 28), cmap='gray') 

ax[4].imshow(images[x], cmap='gray')
ax[4].set_title('Original')  
ax[3].set_title('99% of Explained Variance')
ax[2].set_title('95% of Explained Variance')
ax[1].set_title('90% of Explained Variance')
ax[0].set_title('85% of Explained Variance')   

fig.set_figwidth(18)    
fig.set_figheight(6)

plt.show()
```
